#(Q6_1)
I would like discuss about an example of the Poisson Distribution


Illustration: A statistician records the number of cars that approach a junction. He finds that an average of 2.4 cars  approach the junction every minute.


Explanation:
Assuming the number of cars that approach this junction follows a Poisson distribution, we can discuss the probability that 3 or more cars will approach the junction within a minute.

For this scenario, lamda=2.4. The goal of this Scenario is to find P(X≥3), the probability that there are 3 or more cars approaching the junction within a minute. Since there is no upper limit on the value of k, this probability cannot be computed directly. The complement of P(X≤2), can be computed to give P(X≥3):      

```python
# Implementation through code (Q6_1)
import numpy as np
import scipy.stats as stats

lamda = 2.4
n = np.arange(0,21)
print(n,'\n')

poisson = stats.poisson.cdf(n,lamda)
print('\nCumulative distribution function Data\n')
print(poisson)
PofGreaterthanEquals3cars = 1 - stats.poisson.cdf(2,lamda)
print('\nprobability of greater than 3 cars approaching the junction is\n ',PofGreaterthanEquals3cars)
```

#(Q6_2)
#Another example I would like to discuss is also related to Poisson distribution
A call center receives an average of 3.9 calls every 5 minutes. Each agent can handle one of these calls over the 5 minute period. If a call is received, but no agent is available to take it, then that caller will be placed on hold. Assuming that the calls follow a Poisson distribution, the minimum number of agents needed on duty so that calls are placed on hold at most 10% of the time is to be computed

```python
# Implementation through code (Q6_2)
import numpy as np
import scipy.stats as stats

lamda = 3.9
n = np.arange(0,21)
print(n,'\n')
poisson = stats.poisson.cdf(n,lamda)
print('\nCumulative distribution function Data\n')
print(poisson)
ReqAgentCount = stats.poisson.ppf(0.9,3.9)  # Inverse of "pmf" function evaluated for the success probability of 0.9 to find the number of agents
print('\n Minimum number of agents needed on duty so that calls are placed on hold at most 10% of the time is \n',ReqAgentCount)
```

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
ds = pd.read_csv('/Users/sree/Downloads/PGP_AIML/DataSets/Week4-Stats Project/DS - Part2 - Basketball.csv')

print('\n Order of the dataframe \n',ds.shape)
print ('\n Entry count in the dataframe \n',ds.size)
print('\n\n')
print(ds.head(5))
print('\n\n')
print(ds.tail(5))
print('\n\n')
print(ds.describe())
print('\n\n')
print('\n\n',ds.info(),'\n\n')

# Casting required data to suitable datatypes for analysis purpose

ds['Tournament'] = pd.to_numeric(ds['Tournament'],errors='coerce')
ds['Score'] = pd.to_numeric(ds['Score'], errors='coerce')
ds['PlayedGames'] = pd.to_numeric(ds['PlayedGames'], errors='coerce')
ds['WonGames'] = pd.to_numeric(ds['WonGames'], errors='coerce')
ds['DrawnGames'] = pd.to_numeric(ds['DrawnGames'],errors='coerce')
ds['LostGames'] = pd.to_numeric(ds['LostGames'],errors='coerce')
ds['BasketScored'] = pd.to_numeric(ds['BasketScored'],errors='coerce')
ds['BasketGiven'] = pd.to_numeric(ds['BasketGiven'],errors='coerce')
ds['TournamentChampion'] = pd.to_numeric(ds['TournamentChampion'],errors='coerce')
ds['Runner-up'] = pd.to_numeric(ds['Runner-up'],errors='coerce')
ds['Runner-up'] = pd.to_numeric(ds['Runner-up'],errors='coerce')
ds['HighestPositionHeld']= ds['HighestPositionHeld'].astype(str)
ds['TeamLaunch'] = ds['TeamLaunch'].astype(str)

print(ds.info())
print('\n\n')

print('\n Null values existing column wise\n ',ds.isnull().sum())
print('\n\n')

print(ds.describe(),'\n\n')

#Creating the hybrid features/columns


ds.insert(13,'GameSuccessRate',ds['WonGames']/ds['PlayedGames'])
#print(ds.head(5))
print('\nShape of the ammended dataframe1\n',ds.shape)

ds.insert(14,'GameLosingRate',(ds['LostGames'])/ds['PlayedGames'])
#print(ds.head(5))
print('\nShape of the ammended dataframe2\n',ds.shape)


ds.insert(15,'GameTieRate',(ds['DrawnGames']/ds['PlayedGames']))
#print(ds.head(5))
print('\nShape of the ammended dataframe3\n',ds.shape)


ds.insert(16,'TournamentSuccessRate',ds['TournamentChampion']/ds['Tournament'])
#print(ds.head(5))
print('\nShape of the ammended dataframe4\n',ds.shape)

#Rounding the year in which the team is launched to a single lower bound entry for analysis purpose

ds.insert(17, 'TeamLaunchInt',ds['TeamLaunch'].astype(str).str[0:4])

ds['TeamLaunchInt'] = pd.to_numeric(ds['TeamLaunchInt'],errors='coerce')
#print(ds.info())
#print(ds.head(5))
print('\nShape of the ammended dataframe5\n',ds.shape)

ds.insert(18, 'BasketScoringRate',ds['BasketScored']/(ds['BasketScored']+ds['BasketGiven']))


#print(ds.info())
#print(ds.head(5))
print('\nShape of the ammended dataframe6\n',ds.shape)


ds.insert(19, 'BasketConcedingRate',ds['BasketGiven']/(ds['BasketScored']+ds['BasketGiven']))

#print(ds.info())
#print(ds.head(5))
print('\nShape of the ammended dataframe7\n',ds.shape)


ds.insert(20, 'EventualLosingRate',((ds['GameLosingRate']+ds['GameTieRate'])/(ds['GameLosingRate']+ds['GameTieRate']+ds['GameSuccessRate']))*(1-ds['TournamentSuccessRate']))

#print(ds.info())
#print(ds.head(5))
print('\nShape of the ammended dataframe8\n',ds.shape)


ds.insert(21, 'EventualScoringRate',(ds['Score']/ds['Score'].sum())*(ds['BasketScoringRate']/(ds['BasketScoringRate']+ds['BasketConcedingRate'])))


print(ds.info())
print(ds.head(5))
print('\nShape of the ammended dataframe9\n',ds.shape)


print('\nBest performing team in terms of GameSuccessRate is \n',
      ds[ds['GameSuccessRate']==ds['GameSuccessRate'].max()]['Team'])

print('\nWorst performing team in terms of GameSuccessRate is \n',
      ds[ds['GameLosingRate']==ds['GameLosingRate'].max()]['Team'])

print('\nBest performing team in terms of TournamentSuccessRate is \n',
      ds[ds['TournamentSuccessRate']==ds['TournamentSuccessRate'].max()]['Team'])

print('\nBest performing team in terms of Score is \n',ds[ds['Score']==ds['Score'].max()]['Team'])

print('\nBest performing team in terms of Goals-made(BasketScored) is \n',ds[ds['BasketScored']==ds['BasketScored'].max()]['Team'])

print('\nWorst performing team in terms of Goals-conceded(BasketGiven) is \n',ds[ds['BasketGiven']==ds['BasketGiven'].max()]['Team'])

print('\nOldest team in terms of the year launched \n',ds['TeamLaunchInt'].min())

print('\nYoungest team in terms of the year launched \n',ds['TeamLaunchInt'].max())



print('\nThe best team Company’s management can invest on proposal on managing is \n ',ds[ds['EventualLosingRate']==ds['EventualLosingRate'].min()]['Team'])

print('\nThe company can stick to the following preference while investing in the teams\n\n')

df = ds[['Team','EventualLosingRate','EventualScoringRate']]

df.insert(3,'KeyStatistic',ds['EventualScoringRate']/ds['EventualLosingRate'])

#Remove rows with NaN as the calculated statistic

df.dropna(subset = ['EventualLosingRate'], inplace=True)

df.sort_values(by ='KeyStatistic', axis=0, ascending=False, inplace=True, kind='quicksort', na_position='last')
df.insert(4,'priorityorderhighestfirst',ds['Team'])
print(df)
sns.pairplot(df[['EventualLosingRate','EventualScoringRate','KeyStatistic','priorityorderhighestfirst']],hue="priorityorderhighestfirst")
plt.show()

print('The top recommended team for the company for investment is\n',df[df['KeyStatistic']==df['KeyStatistic'].max()]['Team'])
print('The least recommended team for the company for investment is\n',df[df['KeyStatistic']==df['KeyStatistic'].min()]['Team'])
```
